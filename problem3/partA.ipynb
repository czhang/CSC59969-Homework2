```python
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.io import loadmat
import matplotlib as mpl
import numpy as np

def kelvinToF(temp):
    degreesF = (9.0/5.0)*(temp - 273) + 32
    return degreesF

def loadData(city):
    data = loadmat(city)
    temps= data['temperature']
    temps= [degrees for x in temps for degrees in x]
    F_temps = [kelvinToF(x) for x in temps]
    return F_temps

jburg_all = loadData('Johannesburg.mat')
jburg_f20 = jburg_all[0:20000]
jburg_l20 = jburg_all[len(jburg_all)-20000:]

paris_all = loadData('Paris.mat')
paris_f20 = paris_all[0:20000]
paris_l20 = paris_all[len(paris_all)-20000:]

spaulo_all = loadData('SaoPaulo.mat')
spaulo_f20 = spaulo_all[0:20000]
spaulo_l20 = spaulo_all[len(spaulo_all)-20000:]

sgpore_all = loadData('Singapore.mat')
sgpore_f20 = sgpore_all[0:20000]
sgpore_l20 = sgpore_all[len(sgpore_all)-20000:]

sydney_all = loadData('Sydney.mat')
sydney_f20 = sydney_all[0:20000]
sydney_l20 = sydney_all[len(sydney_all)-20000:]

vcouver_all = loadData('Vancouver.mat')
vcouver_f20 = vcouver_all[0:20000]
vcouver_l20 = vcouver_all[len(vcouver_all)-20000:]

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

citylabels = ['Johannesburg','Paris','SaoPaulo','Singapore','Sydney','Vancouver']
citydata_all = [jburg_all,paris_all,spaulo_all,sgpore_all,sydney_all,vcouver_all]

box = ax.boxplot(citydata_all, notch=True, patch_artist=True)
colors = ['cyan','lightblue','lightgreen','tan','purple','red','yellow']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
x = np.arange(1,7)
plt.xticks(x, citylabels, rotation=45, ha='center', fontsize = 10)
ax.set_xlabel('Cities')
ax.set_ylabel('Temperature (F)')
ax.set_title("World Cities' Temperatures")



    
```
