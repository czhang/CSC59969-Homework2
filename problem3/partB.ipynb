```python
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.io import loadmat
import matplotlib as mpl
import numpy as np
import pandas as pd
from pandas import Series

##########################################################
def kelvinToF(temp):
    degreesF = (9.0/5.0)*(temp - 273) + 32
    return degreesF

def process_data(city):
    data = loadmat(city)
    series = Series(data)
    time_temp = zip(series['time'],series['temperature'])
    
    dec_to_feb = ['12','01','02']
    mar_to_may = ['03','04','05']
    jun_to_aug = ['06','07','08']
    sep_to_nov = ['09','10','11']
    years = {}
    
    yearspan = range(1948,2009)
    for year in yearspan:
        
        fall = []
        winter = []
        spring = []
        summer = []
        seasons = {}
        
        for time, temp  in time_temp:
            if time[0:4] == str(year):
                if time[5:7] in dec_to_feb:
                    winter.append(kelvinToF(temp))
                elif time[5:7] in mar_to_may:
                    spring.append(kelvinToF(temp))
                elif time[5:7] in jun_to_aug:
                    summer.append(kelvinToF(temp))
                elif time[5:7] in sep_to_nov:
                    fall.append(kelvinToF(temp))
            else: continue
                             
        seasons['winter'] = round(np.average(winter), 1)
        seasons['spring'] = round(np.average(spring), 1)
        seasons['summer'] = round(np.average(summer), 1)
        seasons['fall'] = round(np.average(fall), 1)
    
        years[year] = seasons
    
    return years
    
###############################################################


jburg = process_data('Johannesburg.mat')
jburg_winter = [x['winter'] for x in jburg.values()]
jburg_spring = [x['spring'] for x in jburg.values()]
jburg_summer = [x['summer'] for x in jburg.values()]
jburg_fall =   [x['fall'] for x in jburg.values()]

paris = process_data('Paris.mat')
paris_winter = [x['winter'] for x in paris.values()]
paris_spring = [x['spring'] for x in paris.values()]
paris_summer = [x['summer'] for x in paris.values()]
paris_fall =   [x['fall'] for x in paris.values()]

spaulo = process_data('SaoPaulo.mat')
spaulo_winter = [x['winter'] for x in spaulo.values()]
spaulo_spring = [x['spring'] for x in spaulo.values()]
spaulo_summer = [x['summer'] for x in spaulo.values()]
spaulo_fall =   [x['fall'] for x in spaulo.values()]

sgpore = process_data('Singapore.mat')
sgpore_winter = [x['winter'] for x in sgpore.values()]
sgpore_spring = [x['spring'] for x in sgpore.values()]
sgpore_summer = [x['summer'] for x in sgpore.values()]
sgpore_fall =   [x['fall'] for x in sgpore.values()]

sydney = process_data('Sydney.mat')
sydney_winter = [x['winter'] for x in sydney.values()]
sydney_spring = [x['spring'] for x in sydney.values()]
sydney_summer = [x['summer'] for x in sydney.values()]
sydney_fall =   [x['fall'] for x in sydney.values()]

vcouver = process_data('Vancouver.mat')
vcouver_winter = [x['winter'] for x in vcouver.values()]
vcouver_spring = [x['spring'] for x in vcouver.values()]
vcouver_summer = [x['summer'] for x in vcouver.values()]
vcouver_fall =   [x['fall'] for x in vcouver.values()]

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.plot(jburg.keys(), jburg_winter, color = 'tomato', label = 'Johannesburg')
ax.plot(paris.keys(), paris_winter, color = 'steelblue', label = 'Paris')
ax.plot(spaulo.keys(), spaulo_winter, color = 'forestgreen', label = 'Sao Paulo')
ax.plot(sgpore.keys(), sgpore_winter, color = 'maroon', label = 'Singapore')
ax.plot(sydney.keys(), sydney_winter, color = 'goldenrod', label = 'Sydney')
ax.plot(vcouver.keys(), vcouver_winter, color = 'darkblue', label = 'Vancouver')

plt.xticks(jburg.keys(), jburg.keys(), rotation=45, ha='center', fontsize = 7)
ax.set_xlim((1948,2008))
ax.set_ylim((0,100))
ax.set_xlabel('Year')
ax.set_ylabel('Degrees (F)')
ax.set_title('Average Winter Temperatures of Six Cities (1948-2008)')

plt.legend(bbox_to_anchor=(1.05, 1),bbox_transform=plt.gcf().transFigure)
```
