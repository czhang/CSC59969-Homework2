```python
%matplotlib inline
import numpy as np
import matplotlib
#matplotlib.use('Agg')

import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.recfromcsv('crabs.csv')
FL = [rows for rows in data['fl']]
CL = [rows for rows in data['cl']]

fig = plt.figure(figsize=(8,6))

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
gs = mpl.gridspec.GridSpec(1,1)
#put a plot in the first row, first column
ax = fig.add_subplot(gs[0])

#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

#set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

#plot data and customize look of plot
ax.plot(CL,FL, marker='.', color='k', linestyle='',
        markersize=4, clip_on=False)

#set axis limits 
ax.set_xlim((0,50))
ax.set_ylim((0,30))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_xlabel('CL (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('FL (mm)', fontsize=10, family='sans-serif')

ax.set_title('Frontal Lobe Size vs. Carapace Length', fontsize=14, family='sans-serif')

fig.savefig('exercise1-1.png')

    
```

```python
RW = [rows for rows in data['rw']]
BD = [rows for rows in data['bd']]

fig = plt.figure(figsize=(18,6))
plt.subplots_adjust(left=0.125, right=1.0, top=0.9, bottom=0.1, wspace=0.3, hspace=0.5)

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
#put a plot in the first row, first column
ax = fig.add_subplot(1,3,1)
ax2 = ax.twinx() 

ax3 = fig.add_subplot(1,3,2)

#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax3.spines['top'].set_visible(False)
ax3.spines['bottom'].set_linewidth(.5)
ax3.xaxis.set_ticks_position('bottom')

ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_linewidth(0.5)
ax3.yaxis.set_ticks_position('left')



#set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

ax3.xaxis.set_tick_params('major', length=2, labelsize=10)
ax3.yaxis.set_tick_params('major', length=2, labelsize=10)


#plot data and customize look of plot
ax.plot(CL,FL, marker='.', color='red', linestyle='',
        markersize=4, clip_on=False)
ax2.plot(CL,RW, marker ='.', color= 'blue', linestyle = '', markersize=4, clip_on=False)
heatgraph = ax3.scatter(CL,FL,c=RW, cmap='hot')

colbar = plt.colorbar(heatgraph)
colbar.set_label('RW (mm)')

#set axis limits 
ax.set_xlim((0,50))
ax.set_ylim((0,30))
ax2.set_ylim((0,30))
ax3.set_xlim((0,50))
ax3.set_ylim((0,30))


#scale the view to show all datapoints
ax.autoscale_view()
ax3.autoscale_view()

#set labels
ax.set_xlabel('CL (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('FL (mm)', fontsize=10, family='sans-serif', color = 'red')
ax2.set_ylabel('RW (mm)', fontsize= 10, family = 'sans-serif', color = 'blue')
ax3.set_ylabel('FL (mm)', fontsize=10, family='sans-serif')
ax3.set_xlabel('CL (mm)', fontsize=10, family='sans-serif')
fig.text(0.28,0.95,'Frontal Lobe Size & Rear Width vs. Carapace Length', fontsize=14, family='sans-serif')
fig.savefig('exercise1-2.svg')
fig.savefig('exercise1-2.png')


```

```python
CL_MBlue = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
CL_FBlue = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
CL_MOrange = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
CL_FOrange = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

FL_MBlue = [i['fl'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
FL_FBlue = [i['fl'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
FL_MOrange = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
FL_FOrange = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

RW_MBlue = [i['rw'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
RW_FBlue = [i['rw'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
RW_MOrange = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
RW_FOrange = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

fig = plt.figure(figsize=(12,6))

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
#put a plot in the first row, first column
ax = fig.add_subplot(1,1,1)


#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

#set tick properties
ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

#plot data and customize look of plot
RW_MBlue_S=[0.8*n for n in range(len(RW_MBlue))]
RW_FBlue_S=[0.8*n for n in range(len(RW_FBlue))]
RW_MOrange_S=[0.8*n for n in range(len(RW_MOrange))]
RW_FOrange_S=[0.8*n for n in range(len(RW_FOrange))]

ax.scatter(CL_MBlue,FL_MBlue, s=RW_MBlue_S, color='SlateBlue', alpha =.8, label='Blue Male')
ax.scatter(CL_FBlue,FL_FBlue, s=RW_FBlue_S, color='SkyBlue', alpha=.8, label='Blue Female')
ax.scatter(CL_MOrange,FL_MOrange, s=RW_MOrange_S, color='Tomato', alpha=.8, label='Orange Male')
ax.scatter(CL_FOrange,FL_FOrange, s=RW_FOrange_S, color='PeachPuff', alpha=.8, label='Orange Female')

ax.plot(CL_MBlue,FL_MBlue, color='SlateBlue',alpha=.5)
ax.plot(CL_FBlue,FL_FBlue, color='SkyBlue', alpha=.5)
ax.plot(CL_MOrange,FL_MOrange, color='Tomato', alpha=.5)
ax.plot(CL_FOrange,FL_FOrange, color='PeachPuff',alpha = .5)

#set axis limits 
ax.set_xlim((0,50))
ax.set_ylim((0,30))



#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_xlabel('CL (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('FL (mm)', fontsize=10, family='sans-serif')
ax.legend(loc='upper left')
ax.set_title('Frontal Lobe Size & Rear Width  vs. Carapace Length', fontsize=14, family='sans-serif')

fig.savefig('exercise1-4.svg')
fig.savefig('exercise1-4.png')

```
