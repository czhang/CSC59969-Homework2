```python
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.io import loadmat
import matplotlib as mpl
import numpy as np
from scipy.stats.stats import pearsonr

data = loadmat('AQUA.2013089.2055.mat')
spercentage = 0.0045

N_B1 = data['B1'].shape;
N_B6 = data['B6'].shape;

pixels_B1 = N_B1[0] * N_B1[1]
pixels_B6 = N_B6[0] * N_B6[1]

numSamps_B1 = round(pixels_B1*spercentage)
numSamps_B6 = round(pixels_B6*spercentage)

samplesInds_B1 = range(1,pixels_B1)
samplesInds_B6 = range(1,pixels_B6)
np.random.shuffle(samplesInds_B1)
np.random.shuffle(samplesInds_B6)

samplesInds_B1 = samplesInds_B1[0:int(numSamps_B1)]
samplesInds_B6 = samplesInds_B6[0:int(numSamps_B6)]

b1 = data['B1']
sampled_B1 = b1.ravel()[samplesInds_B1]

b6 = data['B6']
sampled_B6 = b6.ravel()[samplesInds_B6]

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.8)
ax.xaxis.set_ticks_position('bottom')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.8)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

z = np.polyfit(sampled_B1, sampled_B6, 2)
p = np.poly1d(z)
slope , yint = np.polyder(p)
corr = pearsonr(sampled_B1, sampled_B6)

ax.plot(sampled_B1, sampled_B6, marker='.', color='black', linestyle='',
        markersize=5, clip_on=False, alpha=0.5)

ax.plot(sampled_B1, p(sampled_B1), marker='.', color='red', linestyle='-',
    markersize=4, clip_on=False, label='Linear Regression Line', alpha=0.5)

ax.set_title('Band 1 Reflectance vs. Band 6 Reflectance', fontsize=14, family='sans-serif')
ax.text(0.45, 0.56, r'$Slope = %s$' % round(slope, 2) , fontsize=12)
ax.text(0.45, 0.52, r'$Correlation Coeff. = %s$' % round(corr[0], 2), fontsize=12)

ax.set_xlabel('Band 1 Reflectance', fontsize=10, family='sans-serif')
ax.set_ylabel('Band 6 Reflectance', fontsize=10, family='sans-serif')

ax.legend()
ax.autoscale_view()

```

```python
N_B7 = data['B7'].shape
pixels_B7 = N_B7[0] * N_B7[1]
numSamps_B7 = round(pixels_B7*spercentage)
samplesInds_B7 = range(1,pixels_B7)
np.random.shuffle(samplesInds_B7)
samplesInds_B7 = samplesInds_B7[0:int(numSamps_B7)]

            
b7 = data['B7']
sampled_B7 = b7.ravel()[samplesInds_B7]

fig2 = plt.figure(figsize=(10,6))
ax2 = fig2.add_subplot(1,1,1)
ax2.spines['top'].set_visible(False)
ax2.spines['bottom'].set_linewidth(.8)
ax2.xaxis.set_ticks_position('bottom')
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_linewidth(0.8)
ax2.yaxis.set_ticks_position('left')
ax2.xaxis.set_tick_params('major', length=2, labelsize=10)
ax2.yaxis.set_tick_params('major', length=2, labelsize=10)

z2 = np.polyfit(sampled_B6, sampled_B7, 2)
p2 = np.poly1d(z2)
slope2 , yint2 = np.polyder(p2)
corr2 = pearsonr(sampled_B6, sampled_B7)



ax2.plot(sampled_B6, sampled_B7, marker='.', color='black', linestyle='',
        markersize=5, clip_on=False, alpha=0.5)

ax2.plot(sampled_B6, p2(sampled_B6), marker='.', color='red', linestyle='-',
    markersize=4, clip_on=False, label='Linear Regression Line', alpha=0.5)

ax2.set_title('Band 6 Reflectance vs. Band 7 Reflectance', fontsize=14, family='sans-serif')
ax2.text(0.03, 0.56, r'$Slope = %s$' % round(slope2, 2) , fontsize=12)
ax2.text(0.03, 0.52, r'$Correlation Coeff. = %s$' % round(corr2[0], 2), fontsize=12)

ax2.set_xlabel('Band 6 Reflectance', fontsize=10, family='sans-serif')
ax2.set_ylabel('Band 7 Reflectance', fontsize=10, family='sans-serif')

ax2.legend()
ax2.autoscale_view()

```

Band 1 and 6 seem to be more related than Band 6 and 7, as they are less spread out.
